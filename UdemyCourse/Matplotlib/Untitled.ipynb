## Introduction to Matplotlib

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot();
```

```python
plt.plot([1,2,3,4]);
```

```python
x = [1,2,3,4]
y = [11, 22, 33 ,44]
plt.plot(x, y);
```

```python
# 1st method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1 , 1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x, y);
```

## Matplotlib example workflow

```python
# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22 ,33 ,44]
# 2.Setup Plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x/label",
      ylabel="y-axis")

# 5. Save and show
fig.savefig('sample-plot.png')
```
